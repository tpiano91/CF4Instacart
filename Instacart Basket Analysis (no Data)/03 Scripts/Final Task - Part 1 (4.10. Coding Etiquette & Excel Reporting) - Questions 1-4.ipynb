### 4. Customer Geography by Region
Question 3: The Instacart officers are interested in comparing customer behavior in different geographic areas. Create a regional segmentation of the data. You’ll need to create a “Region” column based on the “State” column from your customers data set.
##### A) The following wikipedia link lists the US regions that I will be using for this analysis: https://simple.wikipedia.org/wiki/List_of_regions_of_the_United_States
##### B) Determine whether there’s a difference in spending habits between the different U.S. regions. (Hint: You can do this by crossing the variable you just created with the spending flag.)

#### Part A

```python
#checking 'state' values
ords_prods_customer_merge['state'].value_counts(dropna = False)
```

```python
#create 'regions' column according to 'state' and wikipedia designation of each state to a region
regions = []

for value in ords_prods_customer_merge['state']:
    if value in ['Maine','New Hampshire','Vermont','Massachusetts','Rhode Island','Connecticut','New York','Pennsylvania','New Jersey']: 
     regions.append('Northeast')
    elif value in ['Wisconsin','Michigan','Illinois','Indiana','Ohio','North Dakota','South Dakota','Nebraska','Kansas','Minnesota','Iowa','Missouri']: 
     regions.append('Midwest')
    elif value in ['Idaho', 'Montana', 'Wyoming','Nevada','Utah','Colorado','Arizona','New Mexico', 'Alaska', 'Washington','Oregon','Hawaii','California']: 
     regions.append('West')
    else: 
     regions.append('South')
```

```python
ords_prods_customer_merge['regions'] = regions
```

```python
#checking df update
ords_prods_customer_merge.head()
```

```python
#Checking value counts per region
ords_prods_customer_merge['regions'].value_counts()
```

#### Part B

```python
#Create crosstab between 'regions' and 'spending_flag'
crosstab_regions_spendingflag = pd.crosstab(ords_prods_customer_merge['regions'], ords_prods_customer_merge['Spending_Flag'], dropna = False)
```

```python
#Checking crosstab
crosstab_regions_spendingflag
```

#### All regions have many more low spenders than high spenders. "South" region has the highest number of customers and therefore, highest quantity of both high and low spenders. The following list marks the percentage of high/low spenders per region:

Midwest: High spender = 2.05%; Low spender = 97.95%

Northeast: High spender = 1.89%; Low spender = 98.1%

South: High spender = 1.95%; Low spender = 98.05%

West: High spender = 1.94%; Low spender = 98.06%

There is no noticeable difference in the proportions of high/low spenders per region.

### 5. Customer Activity Flag
Question 4: The Instacart CFO isn’t interested in customers who don’t generate much revenue for the app. Create an exclusion flag for low-activity customers (customers with less than 5 orders) and exclude them from the data. Make sure you export this sample.

```python
#Assigning 'activity_flag' based on max_order
ords_prods_customer_merge.loc[ords_prods_customer_merge['max_order'] < 5, 'activity_flag'] = 'Low activity'
ords_prods_customer_merge.loc[(ords_prods_customer_merge['max_order'] >= 5), 'activity_flag'] = 'Regular activity'
```

```python
#Checking value_count of new 'activity_flag'
ords_prods_customer_merge['activity_flag'].value_counts()
```

```python
#Creating subset for active customers
df_active_customers = ords_prods_customer_merge[ords_prods_customer_merge ['activity_flag'] == 'Regular activity' ]
```

```python
#Creating subset for inactive customers
df_inactive_customers = ords_prods_customer_merge[ords_prods_customer_merge ['activity_flag'] == 'Low activity' ]
```

#### 5a. Checking for Mixed Data

```python
for col in customer.columns.tolist():
  weird = (customer[[col]].applymap(type) !=
customer[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customer[weird]) > 0:
    print (col)
```
